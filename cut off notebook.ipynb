```python
import numpy as np
import matplotlib.pyplot as plt
import PIL as pl
import os
import pca
from pca import gaussian_blur_otzu
from skimage.transform import rescale

print('loading files...')
PATH = './runs/run_285'
c=0
images = []
imagenames = []
m=[]
v=[]
for root, dirs, filies in os.walk(PATH):
    for file in filies[:int(len(filies)/2)]:
        image=gaussian_blur_otzu(np.asarray(pl.Image.open(os.path.join(PATH, file))))
        imagename=file.replace('.png','')
        image = 2*(image/255-0.5)
        # images.append(image)
        m.append(np.sum(image))
        v.append(float(imagename.split('_')[-1]))
        # imagenames.append(imagename)
        c+=1
        if c%100==0:
            print(c, 'images added')
print('done! ',c,' images added')
```

```python
%matplotlib tk
plt.figure()
plt.xlim((min(v),max(v)))
plt.ylim((min(m),max(m)))
for x,y in zip(v,m):
    plt.scatter(x,y)
    plt.pause(0.1)
```

```python
import numpy as np
from sklearn.decomposition import PCA
import numpy as np
from sklearn.decomposition import PCA
import cv2
import cv2 as cv
import matplotlib.pyplot as plt
import os


def smart_Gscale(img: np.array) -> np.array:
    # to vector
    V = img.reshape(-1, 3)

    # project to 1D
    pca = PCA(n_components=1).fit(V)
    V = pca.transform(V).reshape(-1)

    # return grey scale matrix
    img = V.reshape(img.shape[:2])
    return np.array(img)
def gaussian_blur_otzu(img):
    # Otsu's thresholding after Gaussian filtering
    blur = cv.GaussianBlur(img, (5, 5), 0)
    ret3, binary = cv.threshold(blur, 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)
    return binary
```
